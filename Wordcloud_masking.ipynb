```
import wikipedia
query = 'Natural Language Processing'
page = wikipedia.page(query).section('Semantics')
x = page.split('\n')
print(x)
print(len(x))
heading = x[::2] # Command /Operation of a list
heading
```

```
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
a = np.array(Image.open("abc.png"))
plt.imshow(a)  # It should be showing an Image
```

```
from wordcloud import WordCloud
mask = np.array(Image.open("abc.png"))
wc = WordCloud(background_color="white", mask=mask)
wc.generate(str(heading))
# create coloring from image
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```
from wordcloud import WordCloud, ImageColorGenerator
mask = np.array(Image.open("abc.png"))
wc = WordCloud(background_color="white", max_words=1000, mask=mask,
               max_font_size=90, random_state=42)
wc.generate(str(heading))
# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[7,7])
plt.imshow(wc.recolor(color_func=image_colors))
plt.axis("off")
plt.show()
```
